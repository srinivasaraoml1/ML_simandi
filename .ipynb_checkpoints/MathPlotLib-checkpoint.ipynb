```python
# Importing matplotlib.pyplot as plt

import matplotlib.pyplot as plt
```

```python
# Line plot

year = [1950, 1970, 1990, 2010]
pop = [2.51, 3.69, 5.23, 6.92]

plt.plot(year, pop)
plt.show()
```

```python
# Scatter plot

year = [1950, 1970, 1990, 2010]
pop = [2.51, 3.69, 5.23, 6.92]

plt.scatter(year, pop)
plt.show()
```
